```python
from itertools import islice
from Bio.Seq import Seq
```

```python
# generator function 
def over_slice(test_str, K):
    itr = iter(test_str)
    res = tuple(islice(itr, K))
    if len(res) == K:
        yield res    
    for ele in itr:
        res = res[1:] + (ele,)
        yield res
```

```python
# calling generator function 
res = [Seq("".join(ele)) for ele in over_slice(rev_str_seq, 15)]
```
